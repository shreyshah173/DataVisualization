```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4]
y = [2,4,6,8]
```

```python
plt.bar(X,y)
```

```python
c = ["r","g","y","b","c"]
```

```python
plt.bar(X,y,color=c)
```

```python
X = ["python","Cpp","Js","C","Java"]
y = [95,100,90,60,2]
plt.bar(X,y,color=c,align="edge",width=0.5,edgecolor="black",linewidth=10,linestyle=":",alpha=0.5)
```

```python
plt.xlabel("Language",fontsize=50)
plt.ylabel("Liked %")
plt.title("Languages vs Liked graph",fontsize=30)
plt.bar(X,y,color=c,label="coding")
plt.legend()
```
