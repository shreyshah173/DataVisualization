```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4]
y = [2,4,6,8]
```

```python
plt.bar(X,y)
```

```python
c = ["r","g","y","b","c"]
```

```python
plt.bar(X,y,color=c)
```

```python
X = ["python","Cpp","Js","C","Java"]
y = [95,100,90,60,2]
plt.bar(X,y,color=c,align="edge",width=0.5,edgecolor="black",linewidth=10,linestyle=":",alpha=0.5)
```

```python
plt.xlabel("Language",fontsize=50)
plt.ylabel("Liked %")
plt.title("Languages vs Liked graph",fontsize=30)
plt.bar(X,y,color=c,label="coding")
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python","c","c++","Java"]
y = [85,70,62,82]
z = [344,76,46,24]
```

```python
width = 0.2
```

```python
p=np.arange(len(x))
p
```

```python
p1 = [j+width for j in p]
p1
```

```python
plt.bar(p,y,width=0.2,color="red")
plt.bar(p1,z,width=0.2,color="y")
```

```python
plt.bar(p,y,width=0.2,color="red")
plt.bar(p1,z,width=0.2,color="y")
plt.xticks(p+width/2,x,rotation=30)
```

```python
plt.barh(p,y,)
plt.bar(p1,z,color="y")
# plt.xticks(p+width/2,x,rotation=30)
```
